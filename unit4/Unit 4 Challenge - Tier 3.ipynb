# Analysis of only "Two Decades"

"which boroughs of London have seen the **greatest increase in housing prices, on average, over the last two decades**?"

Starting here, we are going to consider our time-period of interest to be January 1998 to December 2018 in hopes of matching the rubric so we will redo some of our graphs below.

```python
# Unmelt by pivoting the dates back into columns, selecting only start and end dates: 
newunmelted = cleaned.pivot(
    index='name', columns='month')['price'][['1998-01-01', '2018-12-01']]
# rename: 
newunmelted.columns = ['start_price', 'end_price']
# add a column to show the total change over the entire period: 
newunmelted['absolute_change'] = newunmelted['end_price'] - newunmelted['start_price']
newunmelted.sort_values('absolute_change', inplace=True)
# add another column with the relative change expressed as a percentage
newunmelted['absolute_change'].plot.bar()
# Title the graph
newunmelted['absolute_change'].plot.bar(
    title='Price Changes from Jan 1998 to Dec 2018')
# le6 means in millions so: 
plt.ylabel('Absolute price change in millions')
plt.xlabel('Name of London Borough')
plt.show()
```

We can conclude from this graph that the "boroughs of London that have seen the greatest increase in housing prices, on average, over the last two decades" going back from the end of year 2018 are Kensington & Chelsea, followed by Westminster and Camden.

We can also consider the relative changes, rather than the absolute changes, so let's do that below.

```python
#add a column to show the relative change over the entire period expressed per cent. 
newunmelted['percent_change'] = (newunmelted['absolute_change'] / newunmelted['start_price']) * 100
newunmelted.sort_values('percent_change', inplace=True)
newunmelted['percent_change'].plot.bar(
    title='Relative Price Changes Jan 1998 to Dec 2018')
plt.ylabel('Relative price change per cent (%)')
plt.xlabel('Name of London Borough')
plt.show()
```

We can conclude from the above graph that the greatest relative change in average price occured in the borough of Hackney, followed by Southwark and Waltham Forest. 

For completness, we will redo the relative change graph using the ratios of end price to start price. This should give us the same conclusion as the percentages just did. 

```python
#For completness, we will redo the relative change graph using the ratios of 
#end price to start price. This should give us the same conclusion as the percentages did.
newunmelted['ratio'] = (newunmelted['end_price'] / newunmelted['start_price'])
newunmelted.sort_values('ratio', inplace=True)
newunmelted['ratio'].plot.bar(
    title='ratio of Dec 2018 to Jan 1998 prices')
plt.ylabel('ratio')
plt.xlabel('Name of London Borough')
plt.show()
```

We can see the exact same order above as we did on the preceding graph but this time the y axis is ratio rather than percentage.
